# NatuurSpotter – Package Demonstration

This notebook demonstrates the usage of the **NatuurSpotter** Python package, developed to analyze and visualize moth biodiversity data in **West Flanders (Belgium)** using open ecological observation sources.

**its important to not that Dutch names work better!**
---

## Features:

In this demo, the following core functionalities of the package are showcased:


- **Biodiversity analysis:**:
  - Monthly species richness and diversity metrics
  - Automatic generation of CSV files for Node-RED ingestion

- **Observation mapping:**
  - Interactive HTML map of moth observations in West Flanders
  - Species-based color coding and geographic filtering

- **Seasonal analysis:**
  - Seasonal observation patterns for a selected species
  - Statistical visualization with ecological interpretation

- **Species reporting:**
  - Automated PDF reports including:
    - Species description
    - Rarity status
    - Recent observations table
    - Image sourced from Wikimedia Commons

```python
from natuurspotter import biodiversity_analysis
summary, data = biodiversity_analysis(month=7, year=2025)
print(data.isna)
print(data.describe)
print(data.head())
print(summary)
```
